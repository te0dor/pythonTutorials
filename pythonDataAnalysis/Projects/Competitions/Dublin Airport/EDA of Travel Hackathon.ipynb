```python
import re
missing_values = []
nonumeric_values = []

for column in df:
    # Find all the unique feature values
    uniq = df[column].unique()
    print ("'{}' has {} unique values" .format(column,uniq.size))
    if (uniq.size > 10):
        print("~~Listing up to 10 unique values~~")
    print (uniq[0:10])
    print ("\n-----------------------------------------------------------------------\n")
    
    # Find features with missing values
    if (True in pd.isnull(uniq)):
        s = "{} has {} missing" .format(column, pd.isnull(df[column]).sum())
        missing_values.append(s)
    
    # Find features with non-numeric values
    for i in range (1, np.prod(uniq.shape)):
        if (re.match('nan', str(uniq[i]))):
            break
        if not (re.search('(^\d+\.?\d*$)|(^\d*\.?\d+$)', str(uniq[i]))):
            nonumeric_values.append(column)
            break

print ("Features with missing values:\n{}\n\n" .format(missing_values))
print ("Features with non-numeric values:\n{}" .format(nonumeric_values))
```

### Plotting Bivariate Variables

#### Numerical Variables vs Total

```python
plt.figure(1)
f, axarr = plt.subplots(3, 2, figsize=(6, 5))
total = df.total.values
axarr[0, 0].scatter(df['cat_i_flightno'].values,total)
axarr[0, 0].set_title('cat_i_flightno')
axarr[0, 1].scatter(df.num_flight_weekofyear.values, total)
axarr[0, 1].set_title('num_flight_weekofyear')
axarr[1, 0].scatter(df.num_airport_other_ops.values, total)
axarr[1, 0].set_title('num_airport_other_ops')
axarr[1, 1].scatter(df['num_airport_other_flights'].values, total)
axarr[1, 1].set_title('num_airport_other_flights')
axarr[2, 0].scatter(df.num_city_other_ops.values, total)
axarr[2, 0].set_title('num_city_other_ops')
axarr[2, 1].scatter(df.num_city_other_flights.values, total)
axarr[2, 1].set_title('num_city_other_flights')
f.text(-0.01, 0.5, 'Number of Passengers', va='center', rotation='vertical', fontsize = 18)
plt.tight_layout()
plt.show()
```
